## Transformación

```python
# Generación del campo id 
df_amazon['id'] = 'a' + df_amazon['show_id']
df_disney['id'] = 'd' + df_disney['show_id']
df_hulu['id'] = 'h' + df_hulu['show_id']
df_netflix['id'] = 'n' + df_netflix['show_id']
```

```python
# Unifico los 4 dataframes con la función concat
df = pd.concat([df_amazon, df_disney,df_hulu,df_netflix])
```

```python
# Reemplazo de valores nulos del campo 'rating' por 'G'
df['rating'] = df['rating'].replace(np.nan,'G')
```

```python
# Conversión de fechas al formato AAAA-mm-dd
df['date_added'] = pd.to_datetime(df.date_added)
```

```python
# Conversión de los campos de texto a minúsculas
df = df.applymap(lambda s: s.lower() if type(s) == str else s)
```

```python
# Conversión del campo 'duration' en 2 campos: duration_int (int) y duration_type (min o season) 
df['duration_int'] = df['duration'].str.split(' ').str.get(0)
df['duration_type'] = df['duration'].str.split(' ').str.get(1)
df['duration_type'] = df['duration_type'].replace('seasons','season')
```

Normalizo la columna 'rating'

```python
df['rating'].unique()
```

```python
# reemplazo los datos con la palabra en común 'min' y 'season' por 'sin rating'
df['rating'] = df['rating'].str.replace('\d+\smin','sin rating', regex= True)
df['rating'] = df['rating'].str.replace('\d+\sseason?','sin rating', regex= True)
```

```python
# Normalizo los nombres para 'sin rating'
df['rating'] = df['rating'].str.replace('not_rate','sin rating', regex= True)
df['rating'] = df['rating'].str.replace('not rated','sin rating', regex= True)
df['rating'] = df['rating'].str.replace('unrated','sin rating', regex= True)
```

```python
df['rating'].unique()
```

Añado el promedio de rating (dataset carpeta rating) al dataset 'df'

```python
# Unificamos los dataset de rating
rating1 = pd.read_csv('Datasets/ratings/1.csv')
rating2 = pd.read_csv('Datasets/ratings/2.csv')
rating3 = pd.read_csv('Datasets/ratings/3.csv')
rating4 = pd.read_csv('Datasets/ratings/4.csv')
rating5 = pd.read_csv('Datasets/ratings/5.csv')
rating6 = pd.read_csv('Datasets/ratings/6.csv')
rating7 = pd.read_csv('Datasets/ratings/7.csv')
rating8 = pd.read_csv('Datasets/ratings/8.csv')

rating = pd.concat([rating1,rating2,rating3,rating4,rating5,rating6,rating7,rating8])
```

```python
rating
```

```python
# Cambio el nombre de la columna 'movieId' por 'id'
rating = rating.rename(columns = {'movieId':'id'})

# Agrupo los valores por 'id'
grupos = rating.groupby('id')

# Calculo el promedio de rating por id
rating_promedio = grupos['rating'].mean()
```

```python
# Añado los valores promedio al dataset
df = pd.merge(df,rating_promedio, on = 'id', how = 'left')
```

```python
# Cambio el nombre de la columna 'rating_x' por 'rating' y 'rating_y' por 'score'
df = df.rename(columns = {'rating_x':'rating', 'rating_y':'score'})
```
